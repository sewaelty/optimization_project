```python
import os
import pandas as pd
import numpy as np

project_root = os.path.abspath(os.getcwd())

data_dir = os.path.join(project_root, '..', '..', 'data')

# Summer month

pv_total = pd.read_csv(os.path.join(data_dir, 'Timeseries_2023_PV_Power_Production.csv'), sep=',', skiprows=21)
pv_total
```

```python
# Convert the 'time' column to datetime
pv_total['time'] = pd.to_datetime(pv_total['time'].str[:8] + pv_total['time'].str[9:], format='%Y%m%d%H%M')

# Round to the nearest full hour
pv_total['time'] = pv_total['time'].dt.round('H')

# Display the updated DataFrame
pv_total.head()
```

```python
#change year from 2023 to 2024
pv_total['time'] = pv_total['time'] + pd.DateOffset(years=1)
#change name of column time->timestamp and P-> Energy_production_kWh
pv_total.columns = pv_total.columns.str.replace('time', 'timestamp')
pv_total.columns = pv_total.columns.str.replace('P', 'PV_energy_production_kWh')


# one for august
pv_august = pv_total[(pv_total['timestamp'].dt.month == 8) & (pv_total['timestamp'].dt.year == 2024)]

# one for december
pv_december = pv_total[(pv_total['timestamp'].dt.month == 12) & (pv_total['timestamp'].dt.year == 2024)]
# save the two datasets
pv_august.to_csv(os.path.join(data_dir, 'pv_august.csv'), index=False)
pv_december.to_csv(os.path.join(data_dir, 'pv_december.csv'), index=False)
pv_august
```
